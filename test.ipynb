```python
import subprocess
import os

base_download_path = r"D:\Transmiciones F1"
languages = ["spa", "eng", "fra", "por", "deu", "nld"]
video_track_selector = "0:v:3"

# Iterar sobre cada Gran Premio en el diccionario
for gp_name, sessions in races_dic1.items():
    # Iterar sobre cada sesi√≥n (Carrera, Qualy, etc.)
    for session_name, session_url in sessions.items():
        
        # 1. VERIFICACI√ìN: Saltar si la URL est√° vac√≠a
        if not session_url:
            print(f"--- Saltando {gp_name} - {session_name} (URL no disponible) ---\n")
            continue

        print(f"--- Procesando: {gp_name} - {session_name} ---")

        # 2. PREPARACI√ìN DE RUTAS Y ARCHIVOS
        # Crear la ruta de la carpeta de forma segura
        output_dir = os.path.join(base_download_path, gp_name, session_name)
        os.makedirs(output_dir, exist_ok=True)

        # Nombre base para los archivos de esta sesi√≥n
        file_base_name = f"{gp_name}_{session_name}"
        
        # 3. CONSTRUIR UN √öNICO COMANDO FFMPEG
        cmd = ["ffmpeg", "-i", session_url]

        # Mapear la pista de video a un archivo .mp4
        video_output_path = os.path.join(output_dir, f"{file_base_name}_video.mp4")
        cmd.extend(["-map", video_track_selector, "-c:v", "copy", video_output_path])

        # Mapear CADA pista de audio a su propio archivo .m4a
        for lang in languages:
            audio_output_path = os.path.join(output_dir, f"{file_base_name}_audio_{lang}.m4a")
            cmd.extend([f"-map", f"0:a:m:language:{lang}", "-c:a", "copy", audio_output_path])

        # 4. EJECUTAR EL COMANDO
        print(f"Ejecutando comando para descargar video y todos los audios...")
        print("Comando:", " ".join(cmd)) # Descomenta para depurar y ver el comando completo
        
        result = subprocess.run(cmd, capture_output=True, text=True, encoding='utf-8')

        if result.returncode == 0:
            print(f"‚úÖ Descarga completada exitosamente para {gp_name} - {session_name}\n")
        else:
            print(f"‚ùå Error al procesar {gp_name} - {session_name}.")
            print("FFmpeg stderr (puede indicar qu√© pistas de audio faltan):")
            print(result.stderr)
            print("\n")

print("--- Proceso finalizado para todas las carreras. ---")
```
